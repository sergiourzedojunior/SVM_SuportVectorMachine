# Pipeline

Organização Estruturada:

Os pipelines permitem organizar os passos do processo de aprendizado de máquina de maneira clara e estruturada. Cada etapa do pipeline é nomeada e os passos podem ser facilmente identificados.
Reprodutibilidade:

Ao utilizar pipelines, é mais fácil reproduzir experimentos, pois todos os passos do processo estão encapsulados em uma única estrutura. Isso facilita a replicação do mesmo pipeline em diferentes conjuntos de dados.
Prevenção de Vazamento de Informações:

Pipelines ajudam a evitar vazamento de informações entre conjuntos de treinamento e teste. As transformações (por exemplo, normalização de dados) aprendidas a partir dos dados de treinamento são aplicadas de maneira consistente aos dados de teste.
Facilidade na Validação Cruzada:

Pipelines são compatíveis com a validação cruzada do scikit-learn. Ao realizar a validação cruzada, os passos do pipeline são aplicados corretamente em cada divisão do conjunto de dados, simplificando o processo.
Manutenção e Escalabilidade:

Pipelines tornam mais fácil a manutenção e aprimoramento de modelos. Se você deseja adicionar ou modificar etapas no seu fluxo de trabalho, pode fazer isso de maneira organizada no pipeline, evitando código redundante e facilitando a compreensão do processo.
Ajuste de Hiperparâmetros Unificado:

Quando você precisa ajustar hiperparâmetros, o pipeline simplifica o processo, permitindo ajustar todos os parâmetros de uma vez. Isso é especialmente útil quando há transformações e modelos combinados em uma única estrutura.
Facilidade na Produção:

Pipelines são úteis para levar modelos do estágio de desenvolvimento para a produção. O mesmo pipeline de pré-processamento pode ser usado para transformar novos dados antes de fazer previsões.

# Cross Validation

```python
# Função que retorna a acurácia após fazer um validação cruzada (cross validation)
def Acuracia(clf,X,y):
    resultados = cross_val_predict(clf, X, y, cv=10)
    return metrics.accuracy_score(y,resultados)
```

```python
Acuracia(clf,dataset,classes)
```

```python
# Criando um pipeline
pip_1 = Pipeline([
    ('scaler',StandardScaler()),
    ('clf', svm.SVC())
])
```

```python
# Imprimindo Etapas do Pipeline
pip_1.steps
```

```python
# Chama a função acuracia passando os dados de musicas e as classes
# Usando o pipeline pip_1
Acuracia(pip_1,dataset,classes)
```

```python
# Criando vários Pipelines
pip_2 = Pipeline([
    ('min_max_scaler', MinMaxScaler()),
    ('clf', svm.SVC())
])

pip_3 = Pipeline([
    ('scaler',StandardScaler()),
    ('clf', svm.SVC(kernel='rbf'))
])

pip_4 = Pipeline([
    ('scaler',StandardScaler()),
    ('clf', svm.SVC(kernel='poly'))
])

pip_5 = Pipeline([
    ('scaler',StandardScaler()),
    ('clf', svm.SVC(kernel='linear'))
])
```

```python
# Chama a função acuracia passando os dados de musicas e as classes
# Usando o pipeline pip_2
Acuracia(pip_2,dataset,classes)
```

```python
Acuracia(pip_3,dataset,classes)
```

```python
Acuracia(pip_4,dataset,classes)
```

```python
Acuracia(pip_5,dataset,classes)
```

```python
# Importa o utilitário GridSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
# Lista de Valores de C
lista_C = [0.001, 0.01, 0.1, 1, 10, 100]

# Lista de Valores de gamma
lista_gamma = [0.001, 0.01, 0.1, 1, 10, 100]
```

```python
# Define um dicionário que recebe as listas de parâmetros e valores.
parametros_grid = dict(clf__C=lista_C, clf__gamma=lista_gamma)
```

```python
parametros_grid
```

```python
# Objeto Grid recebe parâmetros de Pipeline, e configurações de cross validation
grid = GridSearchCV(pip_3, parametros_grid, cv=10, scoring='accuracy')
```

```python
# Aplica o gridsearch passando os dados de treino e classes.
grid.fit(dataset,classes)
```
